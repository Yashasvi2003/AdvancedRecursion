```python
import pandas as pd ##pandas for data manipulation
import numpy as np ##numpy for numerical computation
import matplotlib.pyplot as plt ##for data visualization
import seaborn as sns ##seaborn also for data visualization
from sklearn.model_selection import train_test_split ##this function is used to split a dataset into two or more subsets typically training and testing datsets.
from sklearn.linear_model import LinearRegression ##this implements a linear regression model
from sklearn.metrics import mean_squared_error, r2_score
## mean_squared_error:calculates the average of the squares of the errors, which are the differences between the [redicted and actual values. provides the measure of how well the model's predictions match the actuak data.
##r2_score:coefficient of determination which provides a measure of how well the regression meodel explains the variability of the target variable.   R1 score if 1 indicates a prefect fit whereas a score of 0 means the model does not explain any variance in the target variable.
```

```python
data=pd.read_csv('Employee Dataset.csv')## load the dataset
```

```python
data.head() ## view tye firts 5 rows of the dataset: this is by default 5
```

```python
print(data.isnull().sum()) ##check for missing values
```

```python
print(data.describe()) ## display basic statistics
```

```python
print(data.info()) ##display information about the dataset
```

```python
## defining the independent variables(features) and dependent variables( target)
x=data[['active_lifestyle', 'age']] ## these variables will be the features
y=data['salary'] ## this is the target
```

```python
## now split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model= LinearRegression() ## create an instance of the linear regression model
model.fit(x_train, y_train) ## fit the model to the training data
```

```python
y_pred=model.predict(x_test) ## make predictions on the testing data
```

```python
## evaluate the model's performance using the metrics like mean squared error and r-squared R^2
mse=mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

r2=r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
## plotting the actual values and predicted values for visual inspection
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values")
plt.show()
```

```python
## print the coefficient and intercept
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
